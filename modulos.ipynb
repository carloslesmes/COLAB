# Módulos

Los módulos en Python son archivos .py que contienen código reutilizable (funciones, clases, variables) diseñados para organizarse y utilizarse en otros programas mediante la sentencia `import`. Permiten la modularidad, evitando choques de nombres gracias a sus espacios de nombres privados. Python incluye una amplia biblioteca estándar (ej. math, os)

```python
import sys
```

```python
dir(sys)
```

```python
sys.version
```

## Formar de cargar un módulo

1. `import modulo`: Importa todo el módulo. Para usar una función, se utiliza modulo.funcion().
2. `from modulo import funcion`: Importa funciones específicas directamente al espacio de nombres local.
3. `import modulo as alias`: Importa el módulo con un nombre más corto.
4. `from modulo import *`: Importa todo, pero no es recomendable ya que puede causar conflictos de nombres. 

```python
import math
```

```python
dir(math)
```

```python
math.sqrt(math.pi)
```

```python
from math import log
```

```python
log(25)
```
